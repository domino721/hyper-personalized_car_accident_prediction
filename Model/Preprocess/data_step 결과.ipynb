```python
def initialize_resource_paths():
    import os

    # 코랩 mount
    from google.colab import drive
    drive.mount('/content/drive')

    root_directory = '/content/drive/MyDrive/Dataset/'


    """
    원하는 데이터셋은 여기에 추가시키면 돼요
    파일 3개가 뭔차인지 몰라서 이렇게...
    """
    output = dict()
    output['things'] = os.path.join(root_directory, 'things.csv')
    output['things_per'] = os.path.join(root_directory, 'things_per.csv')

    return output

```

```python
resource_paths = initialize_resource_paths()
```

```python
def read_user_enriched(input):
    import pandas as pd

    output = pd.read_csv(input, sep=',', encoding = "cp949", engine='python')

    return output
```

```python
# 불러온 데이터 용량 확인
# 용량이 적당히 크면 잘 불러왔다고  생각가능.
df = read_user_enriched(resource_paths['things'])
per_df = read_user_enriched(resource_paths['things_per'])
# display('RAM memory usage: {} bytes'.format(datasets.memory_usage().sum()))

```

```python
df
```

```python
filtered_df = df[~df['label'].str.contains('e0|e1')]

# accuracy 열을 기준으로 내림차순 정렬
filtered_df = filtered_df.sort_values(by='accuracy', ascending=False)

filtered_df
```

```python
filtered_df2 = per_df[~per_df['label'].str.contains('e0|e1')]

# accuracy 열을 기준으로 내림차순 정렬
filtered_df2 = filtered_df2.sort_values(by='r_square', ascending=False)

filtered_df2
```

```python
filtered_df2 = per_df[~per_df['label'].str.contains('e0|e1')]

# accuracy 열을 기준으로 내림차순 정렬
filtered_df2 = filtered_df2.sort_values(by='mse', ascending=True)

filtered_df2
```
