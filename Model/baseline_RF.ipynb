```python
# RandomForestClassifier 함수
def RF_classifier(X, y):

    # train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y)

    # LogisticRegression 초기화
    rfc = RandomForestClassifier()

    param_grid = {
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
    CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5)
    CV_rfc.fit(X_train, y_train)

    print('Best_params', CV_rfc.best_params_)

    rfc_best = RandomForestClassifier(**CV_rfc.best_params_)
    rfc_best.fit(X_train, y_train)

    # 테스트 데이터셋에 대한 예측 수행
    y_pred = rfc_best.predict(X_test)
    y_proba = rfc_best.predict_proba(X_test)[:, 1]  # AUC 계산을 위한 확률 예측

    # 평가지표 계산
    accuracy = accuracy_score(y_test, y_pred)
    error = 1 - accuracy
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    auc_score = roc_auc_score(y_test, y_proba)

    # 평가지표 출력
    print(f"Accuracy: {accuracy}")
    print(f"Error: {error}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"AUC: {auc_score}")
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
def RF_classifier(X, y):
    # 범주형 변수를 식별
    categorical_features = X.select_dtypes(include=['object']).columns

    # 원-핫 인코더와 RandomForestClassifier를 파이프라인으로 구성
    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', OneHotEncoder(), categorical_features)
        ], remainder='passthrough')

    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', RandomForestClassifier())
    ])

    # 하이퍼파라미터 그리드 설정
    param_grid = {
        'classifier__n_estimators': [200, 500],
        'classifier__max_features': ['auto', 'sqrt', 'log2'],
        'classifier__max_depth': [4, 5, 6, 7, 8],
        'classifier__criterion': ['gini', 'entropy']
    }

    # train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

    # GridSearchCV를 사용한 하이퍼파라미터 튜닝
    CV_rfc = GridSearchCV(pipeline, param_grid=param_grid, cv=5)
    CV_rfc.fit(X_train, y_train)

    # 최적의 파라미터 출력
    print('Best Params:', CV_rfc.best_params_)

    # 테스트 데이터셋에 대한 예측 수행
    y_pred = CV_rfc.predict(X_test)
    y_proba = CV_rfc.predict_proba(X_test)[:, 1]  # AUC 계산을 위한 확률 예측

    # 평가지표 계산
    accuracy = accuracy_score(y_test, y_pred)
    error = 1 - accuracy
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    auc_score = roc_auc_score(y_test, y_proba)

    # 평가지표 출력
    print(f"Accuracy: {accuracy}")
    print(f"Error: {error}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"AUC: {auc_score}")
    return CV_rfc.best_estimator_
```

```python
def RF_classifier_100(X, y):
    # RandomForestClassifier 초기화
    rfc = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)

    # 결과를 저장할 리스트 생성
    accuracies = []
    errors = []
    precisions = []
    recalls = []
    aucs = []

    # 학습 100번 실행
    for i in range(100):
        # train-test split
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

        # 모델 학습
        rfc.fit(X_train, y_train)

        # 예측
        y_pred = rfc.predict(X_test)
        y_proba = rfc.predict_proba(X_test)[:, 1]

        # 평가지표 저장
        accuracies.append(accuracy_score(y_test, y_pred))
        errors.append(1 - accuracy_score(y_test, y_pred))
        precisions.append(precision_score(y_test, y_pred))
        recalls.append(recall_score(y_test, y_pred))
        aucs.append(roc_auc_score(y_test, y_proba))

        # 진행 상황 출력
        print(f"{i + 1}번째 실행 완료")

    # 결과 DataFrame 생성
    results = pd.DataFrame({
        'Accuracy': accuracies,
        'Error Rate': errors,
        'Precision': precisions,
        'Recall': recalls,
        'AUC': aucs
    })

    return results
```

```python
from sklearn.tree import DecisionTreeClassifier

def decision_tree_classifier(X, y):
    # 범주형 변수 식별
    categorical_features = X.select_dtypes(include=['object']).columns

    # 원-핫 인코딩과 DecisionTreeClassifier를 파이프라인으로 구성
    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', OneHotEncoder(), categorical_features)
        ], remainder='passthrough')

    dtc = DecisionTreeClassifier(random_state=42)
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', dtc)
    ])

    # 하이퍼파라미터 그리드 설정
    param_grid = {
        'classifier__max_depth': [3, 5, 7, 10],
        'classifier__min_samples_split': [2, 5, 10],
        'classifier__min_samples_leaf': [1, 2, 4],
        'classifier__criterion': ['gini', 'entropy']
    }

    # GridSearchCV 초기화
    clf = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')

    # 결과를 저장할 리스트 생성
    accuracies = []
    errors = []
    precisions = []
    recalls = []
    aucs = []

    # 학습 100번 실행
    for i in range(100):
        # train-test split
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

        # 모델 학습 (하이퍼파라미터 튜닝 포함)
        clf.fit(X_train, y_train)

        # 최적의 매개변수로 모델 평가
        y_pred = clf.predict(X_test)
        y_proba = clf.predict_proba(X_test)[:, 1]  # 이진 분류인 경우 클래스 확률

        # 평가지표 저장
        accuracies.append(accuracy_score(y_test, y_pred))
        errors.append(1 - accuracy_score(y_test, y_pred))
        precisions.append(precision_score(y_test, y_pred))
        recalls.append(recall_score(y_test, y_pred))
        aucs.append(roc_auc_score(y_test, y_proba))

        # 진행 상황 출력
        print(f"{i + 1}번째 실행 완료: 최적 파라미터 - {clf.best_params_}")

    # 결과 DataFrame 생성
    results = pd.DataFrame({
        'Accuracy': accuracies,
        'Error Rate': errors,
        'Precision': precisions,
        'Recall': recalls,
        'AUC': aucs
    })

    return results
```

```python
from sklearn.neural_network import MLPClassifier

def mlp_classifier(X, y):
    # 범주형 변수 식별
    categorical_features = X.select_dtypes(include=['object']).columns

    # 원-핫 인코더와 MLPClassifier를 파이프라인으로 구성
    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', OneHotEncoder(), categorical_features)
        ], remainder='passthrough')

    mlp = MLPClassifier(max_iter=300)

    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', mlp)
    ])

    # 하이퍼파라미터 그리드 설정
    parameter_space = {
        'classifier__hidden_layer_sizes': [(50,), (100,), (50,50), (100,100)],
        'classifier__alpha': [0.0001, 0.001, 0.01],
        'classifier__learning_rate_init': [0.001, 0.01]
    }

    # GridSearchCV 초기화
    clf = GridSearchCV(pipeline, parameter_space, n_jobs=-1, cv=5, scoring='accuracy')

    # 결과를 저장할 리스트 생성
    accuracies = []
    errors = []
    precisions = []
    recalls = []
    aucs = []

    # 학습 100번 실행
    for i in range(100):
        # train-test split
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

        # 모델 학습 (하이퍼파라미터 튜닝 포함)
        clf.fit(X_train, y_train)

        # 최적의 매개변수로 모델 평가
        y_pred = clf.predict(X_test)
        y_proba = clf.predict_proba(X_test)[:, 1]  # 이진 분류인 경우 클래스 확률

        # 평가지표 저장
        accuracies.append(accuracy_score(y_test, y_pred))
        errors.append(1 - accuracy_score(y_test, y_pred))
        precisions.append(precision_score(y_test, y_pred))
        recalls.append(recall_score(y_test, y_pred))
        aucs.append(roc_auc_score(y_test, y_proba))

        # 진행 상황 출력
        print(f"{i + 1}번째 실행 완료: 최적 파라미터 - {clf.best_params_}")

    # 결과 DataFrame 생성
    results = pd.DataFrame({
        'Accuracy': accuracies,
        'Error Rate': errors,
        'Precision': precisions,
        'Recall': recalls,
        'AUC': aucs
    })

    return results
```
