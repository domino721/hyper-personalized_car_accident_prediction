## Model 함수 정의

```python
from sklearn.model_selection import train_test_split
# from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, roc_auc_score, roc_curve, mean_squared_error, precision_score, recall_score
import math
```

#### 모델 함수

```python
# 로지스틱 회귀 함수
def logistic_classifier(X, y):

    # LogisticRegression 초기화
    model = LogisticRegression()

    # 오차를 저장할 DataFrame 생성
    accuracy = []
    err = []
    precision = []
    recall = []
    auc = []

    # 학습 100번 실행
    for i in range(100):

        # train-test split
        X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2, stratify = y)

        # 모델 학습
        model.fit(X_train, y_train)

        # 예측
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:,1].reshape(-1,1)

        # 평가지표 저장
        accuracy.append(accuracy_score(y_test, y_pred))
        err.append((1 - accuracy_score(y_test, y_pred)))
        precision.append(precision_score(y_test, y_pred))
        recall.append(recall_score(y_test, y_pred))
        auc.append(roc_auc_score(y_test, y_proba))

        # print
        print(f"{i+1}번째 완료")

    result = pd.DataFrame(zip(accuracy, err, auc, precision, recall))
    result.columns = ['accuracy', 'error rate', 'AUC', 'precision', 'recall']

    return result

```

```python
# KNN 함수
from sklearn.neighbors import KNeighborsClassifier

def KNN_classifier(X, y, parameter):


    # best parameter, 오차를 저장할 DataFrame 생성
    neighbor = []

    accuracy = []
    err = []
    precision = []
    recall = []


    for i in range(100):
        # 훈련-검증 train test split
        X_train, X_test, y_train, y_test = train_test_split(X, y,
                                            test_size = 0.2, stratify = y)

        # best param을 저장할 변수 정의
        param = parameter[0]

        # best param 찾기 위해 훈련 검증 test 결과를 저장할 DataFrame 정의
        p_neighbor = []
        p_accuracy = []
        # p_err = []
        # p_precision = []
        p_recall = []

        for j in range(100):

            # 훈련훈련 - 훈련검증 train test split
            X_ttrain, X_valid, y_ttrain, y_valid = train_test_split(X_train, y_train,
                                                test_size = 0.2, stratify = y_train)

            for k, pr in enumerate(parameter):

                # KNN 초기화
                model = KNeighborsClassifier(n_neighbors = pr)

                # param 튜닝을 위한 모델 학습
                model.fit(X_ttrain, y_ttrain)

                # 예측
                y_pred = model.predict(X_valid)

                # 평가지표 저장
                p_accuracy.append(accuracy_score(y_valid, y_pred))
                # p_precision.append(precision_score(y_valid, y_pred))
                # p_recall.append(recall_score(y_valid, y_pred))

                p_neighbor.append(pr)

            # 진행상황 출력
            print(f"{j+1}번째 파라미터 모델 완료")

        # best param 선택을 위해 df 생성
        param_rst = pd.DataFrame(zip(p_neighbor, p_accuracy))   # , err, precision, recall
        param_rst.columns = ['n_neighbors', 'accuracy']     # , 'error rate', 'precision', 'recall'


        # 각 param 별 평균 accuracy 비교를 통해 best param 설정
        bestacc = np.mean(param_rst[param_rst['n_neighbors'] == param]['accuracy'])
        for idx, k in enumerate(parameter[1:]):
            a = np.mean(param_rst[param_rst['n_neighbors'] == k]['accuracy'])
            if (bestacc < a):
                bestacc = a
                param = k

        # best param으로 KNN 초기화
        model = KNeighborsClassifier(n_neighbors = param)

        # 진짜 모델 학습
        model.fit(X_train, y_train)

        # 예측
        y_pred = model.predict(X_test)

        # 평가지표 저장
        accuracy.append(accuracy_score(y_test, y_pred))
        err.append((1-accuracy_score(y_test, y_pred)))
        precision.append(precision_score(y_test, y_pred))
        recall.append(recall_score(y_test, y_pred))

        # 파라미터 저장
        neighbor.append(param)

        # 진행상황 출력
        if ((i+1)%5 == 0):
                print(f"    {i+1}번째 완료")

    result = pd.DataFrame(zip(neighbor, accuracy, err, precision, recall))
    result.columns = ['n_neighbors', 'accuracy', 'error rate', 'precision', 'recall']

    return result

```

### 시각화 함수

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def visual(result):

    plt.figure(figsize=(10, 10))
    plt.boxplot(result, labels = result.columns)
    plt.title("전체")
    plt.show()

    for i, col_name in enumerate(result.columns):
        plt.figure(figsize=(9, 3))

        # 각 column에 대한 boxplot
        sns.boxplot(x = col_name, data = result)
        plt.title(col_name)

        # 그래프 보이기
        plt.show()
```
