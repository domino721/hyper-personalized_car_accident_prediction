```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE



# 오차를 저장할 DataFrame 생성


def count_data_num(data_num_range, model_name, name):

  data_num_range = data_num_range
  num_plots = len(data_num_range)


  all_accuracy = []

  for i, data_num in enumerate(data_num_range):

      accuracy = []

      for j in range(30):

          # model = SVC(kernel = 'linear', C = 1)
          model = model_name


          yn_X_temp = df_encoded.sample(n=data_num)
          # x, y 분리
          yn_X = yn_X_temp.drop(columns = ["사고유무"])
          yn_y = yn_X_temp["사고유무"]


          X_train, X_test, y_train, y_test = train_test_split(yn_X, yn_y ,test_size = 0.2)


          # SMOTE를 이용하여 증강
          smote = SMOTE(sampling_strategy='auto')  # 'auto'는 소수 클래스의 샘플 수를 맞춥니다.
          X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

          # print(len(X_train))



          # 모델 학습
          model.fit(X_train_resampled, y_train_resampled)

          # 예측
          y_pred = model.predict(X_test)

          # 평가지표 저장
          accuracy.append(accuracy_score(y_test, y_pred))

      # print(data_num, " : ", np.median(accuracy))
      all_accuracy.append(accuracy)

      if i % 5 == 0:
        print(i)



  medians = [np.median(arr) for arr in all_accuracy]
  max_values = [np.max(arr) for arr in all_accuracy]
  min_values = [np.min(arr) for arr in all_accuracy]

  plt.figure(figsize=(18, 6))

  # 박스 플롯 그리기
  plt.boxplot(all_accuracy)

  # 중앙값 플롯
  plt.plot(np.arange(1, len(medians) + 1), medians, marker='o', color='r', linestyle='-')

  # 최대값 플롯
  plt.plot(np.arange(1, len(max_values) + 1), max_values, marker='', color='b', linestyle='--')

  # 최솟값 플롯
  plt.plot(np.arange(1, len(min_values) + 1), min_values, marker='', color='b', linestyle='--')

  plt.grid(True)



  # plt.xticks(np.arange(1, len(all_accuracy) + 1), data_num_range)  # 각 boxplot에 대한 이름 설정



  # 그래프 제목과 축 레이블 설정
  plt.title('Accuracy by number of data (' + name + ')' )
  plt.xlabel('number of data (unit: 500)')
  plt.ylabel('accuracy')

  plt.tight_layout()
  # 그래프 표시
  plt.show()

```
