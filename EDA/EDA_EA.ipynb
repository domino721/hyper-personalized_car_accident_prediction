```python
cat_df = df[category_col]
for col_name in category_col :
  if (cat_df[col_name].dtypes == np.float64) | (cat_df[col_name].dtypes == np.int64) :
    cat_df[col_name] = cat_df[col_name].astype(object)

cat_df.info()
```

```python
from scipy.stats import chi2_contingency

chiresult = chi2_contingency(cat_df, correction=False)
print('Chi square: {}'.format(chiresult[0]))
print('P-value: {}'.format(chiresult[1]))
```
